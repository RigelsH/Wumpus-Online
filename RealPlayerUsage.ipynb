```python
import sys
sys.path.append("../")

import wumpus
import json
from RealPlayer import AstarPlayer
import glob
from collections import defaultdict
```

```python
files = glob.glob("data/*")

result = defaultdict()

for file in sorted(files):
    with open(file) as f:
        world_json = json.load(f)

        episode_result = wumpus.run_episode(
            world=wumpus.WumpusWorld.from_JSON(world_json), 
            player=AstarPlayer(name='ASP'), 
            show=False, 
            horizon=1000
        )

        result[file] = episode_result['reward']
```
